```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from ipywidgets import Layout, AppLayout
from IPython.display import display
import functools
 
data = {'year': ['2000', '2000','2000','2000','2001','2001','2001','2001', '2002',  '2002', '2002', '2002',
                 '2003','2003','2003','2003','2004', '2004','2004','2004', '2005', '2005', '2005', '2005', 
                 '2006', '2006', '2006', '2006', '2006', '2007', '2007', '2007', '2007', '2008', '2008', '2008', '2008',
                 '2009', '2009', '2009', '2009'],
        
        'purpose':['Holiday', 'Business', 'VFR', 'Study', 'Holiday', 'Business', 'VFR', 'Study', 'Holiday', 'Business',
                   'VFR', 'Study', 'Holiday', 'Business', 'VFR', 'Study', 'Holiday', 'Business', 'VFR', 'Study', 'Holiday',
                   'Business', 'VFR', 'Study', 'Holiday', 'Business', 'VFR', 'Study', 'Holiday', 'Business', 'VFR', 'Study',
                   'Holiday', 'Business', 'VFR', 'Study', 'Holiday', 'Business', 'VFR', 'Study', 'Holiday'
                  ], 
                'market':['Belgium', 'Luxembourg', 'France', 'Spain', 'Norway', 'Sweden', 'Germany', 'Austria', 'Denmark',
                  'Portugal', 'Greece', 'Croatia', 'Belgium', 'Luxembourg', 'France', 'Spain', 'Norway', 'Sweden',
                  'Germany', 'Austria', 'Denmark', 'Portugal', 'Greece', 'Croatia', 'Belgium', 'Luxembourg', 'France',
                  'Spain', 'Norway', 'Sweden', 'Germany', 'Austria', 'Denmark', 'Portugal', 'Greece', 'Croatia', 'Belgium',
                  'Luxembourg', 'France', 'Spain', 'Norway'
                 ]} 
 
df_london = pd.DataFrame (data, columns = ['year','purpose', 'market'])
 
output_dataframe = None

# Get our unique values
ALL = 'ALL'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique

output = widgets.Output()

# Dropdown listbox
dropdown_year = widgets.Dropdown(description='Year',
                                  options = unique_sorted_values_plus_ALL(df_london.year))

        
        
# Function to filter our dropdown listboxe
def common_filtering(year):
    global output_dataframe
    df = df_london.copy()
        
    filters = []
    
    # Evaluate our dropdown listbox and return booleans for our selections
    if year is not ALL:
        filters.append(df['year'] == year)
    
    output.clear_output()
    
    with output:
        if filters:
            df_filter = functools.reduce(lambda x,y: x&y, filters)
            output_dataframe = df.loc[df_filter]
        else:
            output_dataframe = df
        display(output_dataframe)

def dropdown_year_eventhandler(change):
    common_filtering(change.new)

dropdown_year.observe(dropdown_year_eventhandler, names='value')

ui = widgets.HBox([dropdown_year])

display(ui, output)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Example dataset
data = pd.DataFrame({
    'Name': ['John', 'Emily', 'Michael', 'Jessica', 'David'],
    'Age': [25, 30, 28, 32, 27],
    'Score': [85, 92, 88, 90, 87],
    'City': ['New York', 'London', 'Paris', 'New York', 'Tokyo']
})

# Create the integer slider widgets
age_slider = widgets.IntSlider(
    value=25,
    min=0,
    max=100,
    step=1,
    description='Age:'
)

score_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=100,
    step=1,
    description='Score:'
)

# Create the text input widget for City filtering
city_text = widgets.Text(
    placeholder='Enter city...',
    description='City:'
)

# Create the toggle buttons for logical conditions
condition_toggle_1 = widgets.ToggleButtons(
    options=['AND', 'OR'],
    description='Condition 1:'
)

condition_toggle_2 = widgets.ToggleButtons(
    options=['AND', 'OR'],
    description='Condition 2:'
)

# Define the filtering function
def filter_dataset(age, score, city, condition_1, condition_2):
    if condition_1 == 'AND':
        filter_1 = (data['Age'] >= age) & (data['Score'] >= score)
    else:
        filter_1 = (data['Age'] >= age) | (data['Score'] >= score)
    
    if condition_2 == 'AND':
        filter_2 = data['City'].str.contains(city, case=False)
    else:
        filter_2 = ~data['City'].str.contains(city, case=False)
    
    filtered_data = data[filter_1 & filter_2]
    return filtered_data

# Create the button widget
button = widgets.Button(description='Filter')

# Define the button click event
def on_button_click(b):
    age = age_slider.value
    score = score_slider.value
    city = city_text.value
    condition_1 = condition_toggle_1.value
    condition_2 = condition_toggle_2.value
    filtered_data = filter_dataset(age, score, city, condition_1, condition_2)
    print(filtered_data)  # Access the filtered dataset here

# Attach the button click event
button.on_click(on_button_click)

# Display the widgets
display(widgets.VBox([age_slider, score_slider, condition_toggle_1, city_text, condition_toggle_2, button]))
```
